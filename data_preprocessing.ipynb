```python
from datasets import load_dataset
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re

ds = load_dataset("gopalkalpande/bbc-news-summary")

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess(text):
    lemmatizer = WordNetLemmatizer()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    text = text.split()
    text = [lemmatizer.lemmatize(word) for word in text if word not in set(stopwords.words('english'))]
    text = ' '.join(text)
    return text

for i in range(len(ds['train'])):
    ds['train'][i]['Articles'] = preprocess(ds['train'][i]['Articles'])
    print("row", i, "done")

print(ds['train'][1522]['Articles'])
```
